````python
import json
import os

# Specify the folder path
folder_path = 'ojk-all/data_true_pdf/'

# List all files and folders in the directory
all_files = os.listdir(folder_path)

# Filter out only files
DOCUMENT_NAMES = [file for file in all_files if os.path.isfile(os.path.join(folder_path, file))]

f = open("dic_tracing_ojk_new.json")
dic = json.load(f)
````

````python
import pandas as pd

df = pd.read_csv("ojk-all/files_metadata.csv")
df.head()
````

````python
import pymupdf

def find_trace(file_path):
    doc = pymupdf.open(file_path)
    reg_number = df[df["new_filename"] == file_path.split('/')[-1]]['nomor_regulasi'].values[0]
    full_text = ""
    for page in doc:
        text = page.get_text()
        full_text += text
    
    return full_text, reg_number

def find_trace_ojk(doc_idx):
    return find_trace('ojk-all/data_true_pdf/{}'.format(DOCUMENT_NAMES[doc_idx]))
````

## LLM

````python
from typing import List

from langchain.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.pydantic_v1 import BaseModel, Field, validator
from langchain_openai import ChatOpenAI
````

````python
from dotenv import load_dotenv
load_dotenv()
model = ChatOpenAI(temperature=0, model="gpt-4o-mini")
````

````python
PROMPT_TEMPLATE = """
Extract and categorize the regulation numbers mentioned in the text. Specifically, identify and list the regulations that have been *repealed* and those that *repeal* other regulations.

**Important Note:**
When analyzing the text, pay special attention to keywords like "Pasal," "Ayat," or "huruf" that may appear prior to a regulation number. These keywords indicate that only specific parts (such as chapters, sections, or paragraphs) of the regulation are affected. In such cases, do **not** include these regulation numbers in the list of repealed regulations, as other parts of these regulations still remain in force. The focus should be on entire regulations that are either fully repealed or that fully repeal other regulations.

==========================================

**Example Input Context:**

Peraturan yang mencabut: 62/POJK.03/2020 
 BAB XII 
KETENTUAN PENUTUP 
 
Pasal 152 
 
Pada saat Peraturan Otoritas Jasa Keuangan ini mulai berlaku: 
a. Peraturan Otoritas Jasa Keuangan Nomor 20/POJK.03/2014 tentang Bank Perkreditan Rakyat;
b. Surat Edaran Otoritas Jasa Keuangan Nomor 16/SEOJK.03/2015 tentang Bank Perkreditan Rakyat; 
c. Pasal 11 sampai dengan Pasal 19 Peraturan Otoritas Jasa Keuangan Nomor 12/POJK.03/2016 tentang Kegiatan Usaha dan Wilayah Jaringan Kantor Bank Perkreditan Rakyat Berdasarkan Modal Inti (Lembaran Negara Republik Indonesia Tahun 2016 Nomor 34, Tambahan Lembaran Negara Republik Indonesia Nomor 5849); 
d. Peraturan Bank Indonesia Nomor 10/9/PBI/2008 tentang  Perubahan Izin Usaha Bank Umum menjadi Izin Usaha Bank Perkreditan Rakyat dalam Rangka Konsolidasi (Lembaran Negara Republik Indonesia Tahun 2008 Nomor 35, Tambahan Lembaran Negara Republik Indonesia Nomor 4823); dan 
e. Surat Edaran Bank Indonesia Nomor 12/36/DPNP tentang Perubahan Izin Usaha Bank Umum menjadi Izin Usaha Bank Perkreditan Rakyat secara Mandatory dalam rangka Konsolidasi, dicabut dan dinyatakan tidak berlaku.
```

==========================================

**Expected Thinking:**
Remember, if you encounter keywords such as "Pasal," "Ayat," or "huruf" before a regulation number, exclude it from the list of repealed regulations since it suggests that only a portion of the regulation is affected, and the remainder is still valid.
In the example, I do not include 12/POJK.03/2016 as only some chapters that are changed.
```plaintext
1. **Regulation Repealed:** [20/POJK.03/2014, 16/SEOJK.03/2015, 10/9/PBI/2008, 12/36/DPNP]
2. **Regulation Repealing:** [62/POJK.03/2020]
```

==========================================

The example above is only for illustrative purposes to guide your thinking. Now, you need to apply this process to the following context:
{context}

The output format instruction is:
{format_instructions}
```

"""
````

````python
class Peraturan(BaseModel):
    mencabut: str = Field(description="Nomor Peraturan yang mencabut")
    dicabut: list[str] = Field(description="Semua Nomor Peraturan yang dicabut secara Keseluruhan dan bukan beberapa pasal atau ayat tertentu saja")

parser = PydanticOutputParser(pydantic_object=Peraturan)

prompt = PromptTemplate(
    template= PROMPT_TEMPLATE,
    input_variables=["context"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

chain = prompt | model | parser
````

````python
def extract_text(idx):
    range_text = dic[str(idx)]
    text, reg = find_trace_ojk(idx)
    selected = text[range_text[0]:range_text[1]]
    full_text = f"Peraturan yang mencabut: {reg} \n {selected}"
    return full_text
````

## Iterate all

````python
result = {}
````

````python
for key, val in dic.items():
    temp = chain.invoke({"context": extract_text(int(key))})
    result[key] = temp
    print("Success for index", key)
````

````python
def custom_encoder(obj):
    if isinstance(obj, Peraturan):
        return {"Mencabut": obj.mencabut, "Dicabut": obj.dicabut}
    # Let the base class default method raise the TypeError
    raise TypeError(f"Object of type {type(obj).__name__} is not JSON serializable")

def custom_decoder(dct):
    if "Mencabut" in dct:
        return Peraturan(mencabut=dct["Mencabut"], dicabut=dct["Dicabut"])
    return dct

with open('result_ojk.json', 'w') as f:
    json.dump(result, f, default=custom_encoder)
````
