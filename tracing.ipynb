```python
def find_trace(file_path):
    full_text = ""
    doc = pymupdf.open(file_path)
    for page in doc:
        text = page.get_text()
        full_text += text
    
    return full_text

def find_trace_ojk(doc_idx):
    return find_trace('ojk-all/data_true_pdf/{}'.format(DOCUMENT_NAMES[doc_idx]))
```

```python
find_trace_ojk(2).find("dicabut dan dinyatakan tidak berlaku")
```

```python
print(len(DOCUMENT_NAMES))
```

```python
doc_dicabut_idx = {}

for i in range(len(DOCUMENT_NAMES)):
    idx = find_trace_ojk(i).find("dicabut dan dinyatakan tidak berlaku") 
    if idx != -1:
        doc_dicabut_idx[i] = idx
        print("There are document dicabut by the peraturan index", i)
```

```python
doc_dicabut_end = {}
```

```python
for id, idx in doc_dicabut_idx.items():
    txt = find_trace_ojk(id)
    target_idx = txt.rfind("BAB", 0, idx)
    doc_dicabut_end[id] = target_idx
    print("Done for document", id)
```

```python
find_trace_ojk(24)
```

```python
dic_res = {}

for id, idx in doc_dicabut_idx.items():
    dic_res[id] = [doc_dicabut_end[id], idx]
```

```python
filtered_dic_res = {}

for key, val in dic_res.items():
    if val[0] != -1 and (not(val[1]-val[0] > 5000)):
        filtered_dic_res[key] = [val[0], val[1]+37]
```
