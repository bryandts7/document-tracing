```python
import os

def list_pdf_files(parent_folder):
    # List to store the relative paths of PDF files
    pdf_files = []

    # Walk through the directory tree
    for root, dirs, files in os.walk(parent_folder):
        for file in files:
            if file.endswith('.pdf'):
                # Get the relative path to the file from the parent folder
                relative_path = os.path.relpath(os.path.join(root, file), parent_folder)
                # Replace backslashes with forward slashes
                relative_path = relative_path.replace(os.sep, '/')
                full_path = os.path.join(parent_folder, relative_path).replace(os.sep, '/')
                pdf_files.append(full_path)
    
    return pdf_files

# Usage
parent_folder = 'BI_Docs'  # Replace with your parent folder path
pdf_files = list_pdf_files(parent_folder)
print(len(pdf_files))
# Print out the list of PDF files
for pdf in pdf_files:
    print(pdf)
```

```python
from documents_extract_ojk import extract_text_from_pdf
from paddleocr import PaddleOCR


ocr_id = PaddleOCR(use_angle_cls=True, lang='id', show_log=False, use_gpu=True)
ocr_en = PaddleOCR(use_angle_cls=True, lang='en', show_log=False, use_gpu=True)


for pdf in pdf_files:
    try:
        if pdf not in res:
            if pdf[18:20] == "en":
                text = extract_text_from_pdf(pdf, ocr_en, 0.98)
            else:
                text = extract_text_from_pdf(pdf, ocr_id, 0.98)

            print("Successfully reading file", pdf)
            res[pdf] = text
        else:
            print(pdf, "already available")
    except:
        print("Failed to process", pdf)
        continue
```
